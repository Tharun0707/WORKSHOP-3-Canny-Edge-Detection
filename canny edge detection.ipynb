```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("dog.png")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
```

```python
blur_img=cv2.blur(img,(2,2))
plt.figure()
plt.imshow(cv2.cvtColor(blur_img, cv2.COLOR_BGR2RGB))
plt.title("Blurred Image")
plt.show()
```

```python
median=np.median(blur_img)
lower_threshold=int(max(0,0.2*median))
upper_threshold=int(min(255,1/3*median))
edges=cv2.Canny(blur_img,lower_threshold,upper_threshold)
plt.figure()
plt.imshow(edges,cmap='gray')
plt.title("Edges detected using Canny")
plt.show()
```
